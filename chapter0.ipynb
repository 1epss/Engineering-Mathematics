```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import sympy
```

```python
# 함수값 구하기

# 1. f(x) =  5x^2 + 7x -6, f(4) = ?
x = sympy.symbols('x')
f = 5*x**2 + 7*x -6
print('1 :', f.subs(x,4))

# 2. f(x) = e^x + 3^x lnx, f(e) = ?
x = sympy.symbols('x')
f = sympy.exp(x) + 3**x*sympy.ln(x)
print('2 :', f.subs(x,sympy.E))

# 3. f(x) = sin x + (cos x / x), f(pi) = ?
x = sympy.symbols('x')
f = sympy.sin(x) + (sympy.cos(x) / x)
print('3 :', f.subs(x, sympy.pi))

# 4. f(x,y) = (x^2+y^2)^1/2, f(-2,5) = ?
x = sympy.symbols('x')
y = sympy.symbols('y')
f = sympy.sqrt(x**2 + y**2)
print('4 :', f.subs([(x,-2),(y,5)]))
```

```python
# 그래프 그리기

# 5. f(x) = x^2
x = sympy.symbols('x')
f = x**2
p1 = sympy.plot(f,(x,-2,2), show=False)

# 6. g(x) = e^x
x = sympy.symbols('x')
g = sympy.E**x
p2 = sympy.plot(g,(x,-2,2), show=False)

# 7. h(x) = ln(x)
x = sympy.symbols('x')
h = sympy.ln(x)
p3 = sympy.plot(h,(x,0.1,10), xlim=(-1,10) ,ylim=(-5,5), show=False)

# 8. i(x) = sin(x)
x = sympy.symbols('x')
i = sympy.sin(x)
p4 = sympy.plot(i,(x,-sympy.pi,sympy.pi), xlim=(-2*sympy.pi,2*sympy.pi) ,ylim=(-1,1), show=False)

# 한번에 보기쉽게 plotgrid로 그래프 정렬, matplotlib의 subplot과 비슷한 느낌
# plotting.Plotgrid(row, column, *args, show=True, size=None, **kwargs)
plotgrid = sympy.plotting.PlotGrid(1, 4, p1, p2, p3, p4, size=(13, 5))
```

```python
# graphs for various functions

# 9. x = cos(t), y = sin(t) (t = 0<=t<=2pi)
# 매개변수 t로 정의된 함수의 그래프 - plot_parametric()
t = sympy.symbols('t')
x = sympy.cos(t)
y = sympy.sin(t)
p1 = sympy.plot_parametric(x,y,(t,0,2*sympy.pi), xlim=(-3,3), ylim=(-3,3), show=False)

# 10. x^2+2y^2 = 10
# 음함수 그래프 - plot_implicit()
x = sympy.symbols('x')
y = sympy.symbols('y')
f = sympy.Eq(x**2+2*y**2, 10) 
p2 = sympy.plot_implicit(f, show=False)

# 11. f(x,y) = x / e^(x^2+y^2)
# 다변수 함수 - plotting.plot3d()
x = sympy.symbols('x')
y = sympy.symbols('y')
f = x / sympy.exp(x**2+y**2)
p3 = sympy.plotting.plot3d(f,(x, -3,3),(y,-3,3), show=False)

# 12. x = sin(t), y = cos(t), z =t (t = 0<=t<=2pi)
# 3차원 공간에서 매개변수 t로 정의된 함수의 그래프 - plotting.plot3d_parametric_line()
t = sympy.symbols('t')
x = sympy.sin(t)
y = sympy.cos(t)
z = t
p4 = sympy.plotting.plot3d_parametric_line(x, y, z, (t, 0, 2*sympy.pi), show = False)

plotgrid = sympy.plotting.PlotGrid(1, 4, p1, p2, p3, p4, size=(15, 4))

```

```python
# 13. f(x) = 5x^2 + 7x -6의 도함수, 이계도함수
x = sympy.symbols('x')
f = 5*x**2 + 7*x -6
print('13 :', sympy.diff(f,x))
print('13-1 :', sympy.diff(f,x,x))

# 14. f(x) = sin x + (cos x / x)의 도함수
x = sympy.symbols('x')
f = sympy.sin(x) + (sympy.cos(x) / x)
print('14 :', sympy.diff(f,x))

# 15. f(x,y) = sqrt(x^2 + y^2) y에 대해 편미분, x,y 이계편도함수 
x = sympy.symbols('x')
y = sympy.symbols('y')
f = sympy.sqrt(x**2 + y**2)
print('15 :', sympy.diff(f,y))
print('15-1 :', sympy.diff(f,x,y))

# 16. f(x,y) = x/y, (x,y) = (1,-1) 에서의 x,y 이계편도함수
x = sympy.symbols('x')
y = sympy.symbols('y')
f = x/y
print('16 :', sympy.diff(f,x,y).subs([(x,1),(y,-1)]))
```

```python
# 17. x sin y + y cos x = 0, dy/dx
# 음함수의 도함수 & 편도함수 - sympy.idiff()
x = sympy.symbols('x')
y = sympy.symbols('y')
f = x*sympy.sin(y) + y*sympy.cos(x)
print('17 :', sympy.idiff(f,y,x))

# 18. x^3 + y^3 + z^3 +3xyz = 0, ∂z/∂y
x = sympy.symbols('x')
y = sympy.symbols('y')
z = sympy.symbols('z')
f = x**3 + y**3 + z**3 + 3*x*y*z
print('18 :', sympy.idiff(f,z,y)) 

# 19. z = xy^2, x = t^3, y = t^4, dz/dt
t = sympy.symbols('t')
x = t**3
y = t**4
z = x*y**2
print('19 :', sympy.diff(z,t))

# 20. z = u^3 + v^3, u = x + y, v = x - y, ∂z/∂x
x = sympy.symbols('x')
y = sympy.symbols('y')
u = x + y
v = x - y
z = u**3 + v**3
print('20 :', sympy.diff(z,x).simplify())
```

```python
# 부정적분 계산하기

# 21. f(x) = 5x^2 + 7x -6 의 부정적분 f(x) dx
x = sympy.symbols('x')
f = 5*x**2 + 7*x -6
print('21 :', sympy.integrate(f,x))

# 22. f(x) = x sqrt(2x+1) 의 부정적분 f(x) dx
x = sympy.symbols('x')
f = x * sympy.sqrt(2*x + 1)
print('22 :', sympy.integrate(f,x).simplify())

# 23. f(x) = 2e^-x + 3x^(x+1) 의 부정적분 f(x) dx
x = sympy.symbols('x')
f = 2*sympy.exp(-x) + 3**(x+1)
print('23 :', sympy.integrate(f,x))

# 24. f(x) = sin^2(x/2) 의 부정적분 f(x) dx
x = sympy.symbols('x')
f = sympy.sin(x/2) ** 2
print('24 :', sympy.integrate(f,x).simplify())
```

```python
# 정적분 계산하기

# 25. ∫4_2 1 / (x*2 - x) dx
x = sympy.symbols('x')
f = 1 / (x**2 - x)
print('25 :', sympy.integrate(f,(x,2,4)).simplify())

# 26. ∫pi_0 x cos x dx
x = sympy.symbols('x')
f = x * sympy.cos(x)
print('26 :', sympy.integrate(f,(x,0,sympy.pi)))

# 27. ∫1_0 xe^-x dx
x = sympy.symbols('x')
f = x * sympy.exp(-x)
print('27 :', sympy.integrate(f,(x,0,1)))

# 28. ∫infinity_1 1/sqrt(x) dx
x = sympy.symbols('x')
f = 1 / sympy.sqrt(x)
print('28 :', sympy.integrate(f,(x,1,sympy.oo)))
```

```python
# 다중적분 계산하기

# 29. ∫2_1 ∫5-y_y e^(x+3y) dxdy
x = sympy.symbols('x')
y = sympy.symbols('y')
f = sympy.exp(x+3*y)
print('29 :', sympy.integrate(f,(x,y,5-y),(y,1,2)))

# 30. ∫2_0 ∫2-x_0 ∫2-x-y_0 dzdydx
x = sympy.symbols('x')
y = sympy.symbols('y')
z = sympy.symbols('z')
f = 1
print('30 :', sympy.integrate(f,(z,0,2-x-y),(y,0,2-x),(x,0,2)))

# 31. 원점 = 중심, 반경 = 1인 원의 내부 중 1사분면에 놓여있는 영역 R에 대하여, ∫ ∫_R sqrt(x^2 + y^2) dA
# 극좌표계를 이용한 계산, dA = rd rdθ
r = sympy.symbols('r')
theta = sympy.symbols('theta')
x = r * sympy.cos(theta)
y = r * sympy.sin(theta)
f = sympy.sqrt(x**2 + y**2)
print('31 :', sympy.integrate(f*r,(r,0,1),(theta,0,sympy.pi/2)))

# 32. 원점 = 중심, 반경 = 4인 원의 내부 영역 R에 대하여, ∫ ∫_R sqrt(4x^2 + 4y^2 + 1) dA
r = sympy.symbols('r')
theta = sympy.symbols('theta')
x = r * sympy.cos(theta)
y = r * sympy.sin(theta)
f = sympy.sqrt(4*x**2 + 4*y**2 + 1)
print('32 :', sympy.integrate(f*r,(r,0,4),(theta,0,sympy.pi * 2)).simplify())
```

```python
# 대수방정식의 해 계산

# 33. 이차방정식 x^2 + 2 = 3x의 해
x = sympy.symbols('x')
print('33 :', sympy.solve(sympy.Eq(x**2 + 2, 3*x), x))

# 34. 이차방정식 ax^2 + bx + c = 0의 해
x = sympy.symbols('x')
a = sympy.symbols('a')
b = sympy.symbols('b')
c = sympy.symbols('c')
print('34 :', sympy.solve(sympy.Eq(a*x**2 + b*x + c, 0), x))

# 35. 지수방정식 e^x = 5의 해
x = sympy.symbols('x')
print('35 :', sympy.solve(sympy.Eq(sympy.exp(x), 5), x))
```
